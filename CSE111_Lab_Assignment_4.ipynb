```python
#todo
class Customer:
  
  def __init__(self,name):
    self.name = name
  
  def greet(self,greet_name=None):
    if ( greet_name==None):
      print('Hello!')
    else:
      print('Hello '+greet_name+"!")
  
  def purchase(self,*items):
    print("{}, you purchased {} item(s):".format(self.name,len(items)))
    for i in items:
      print(i)



customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#todo
class TaxiLagbe:
     def __init__(self,texi,area):
         self.texi = texi
         self.area = area
         self.p_dict={}
     def addPassenger(self,*args):
         self.fare = 0
         if len(self.p_dict) == 4:
            print("Taxi Full! No more passenger can be added!")
            return None
         for i in args:
            self.p_dict[i.split("_")[0]]=int(i.split("_")[1])
         for j in args:
            print(f'Dear {j.split("_")[0]}! Welcome to TaxiLagbe.')
         for x in self.p_dict.values():
             self.fare+=x
     def printDetails(self):
         print(f"""Trip info for Taxi number:{self.texi}
               \nThis taxi can cover only{self.area} area
               \nPassenger lists:""")
         passenger = ", ".join(self.p_dict.keys())
         print(passenger)

#########################################################################

taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
                
           
             
```

```python
#todo
class StudentDatabase:
     def __init__(self,name,id):
         self.name = name
         self.id = id
         self.grades = {}

     def calculateGPA(self,course,sem):
         self.course = course
         self.sem = sem
         gpa = 0
         c_list=[]         
         for i in self.course:
             c_list.append(i.split(":")[0])
             gpa+=float(i.split(":")[1])
         cgpa = gpa/len(c_list)
         self.grades[self.sem]={tuple(c_list):cgpa}
         
     def printDetails(self):
         print(f'Name: {self.name} \nID: {self.id}')
         for i in self.grades:
            print(f'Courses taken in {i}:')
            for j in self.grades[i].keys():
              for k in j:
                print(k)
            for l in self.grades[i].values():
              print(f'GPA: {l}')
                
################################################################################
             
s1 = StudentDatabase('Pietro', '10101222' )
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
